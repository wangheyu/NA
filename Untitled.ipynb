```python
import sympy as sym         # 调入符号计算库
from sympy import symbols   # 符号变量声明
from sympy.plotting import plot
x = symbols('x')            # x 是符号变量
```

```python
f = sym.sin(sym.pi * x)
f
```

```python
phi = sym.zeros(1, 3)
phi
```

```python
phi[0] = 1
phi[1] = x - sym.Rational(1,2)
phi[2] = x**2 - x + sym.Rational(1,6)
phi
```

```python
G = sym.zeros(3)
for i in range(3):
    for j in range(3):
        # 顺便验证了正交性
        G[i, j] = sym.integrate(phi[i] * phi[j], (x, 0, 1))
G
```

```python
rhs = sym.zeros(3, 1)
for i in range(3):
    rhs[i] = sym.integrate(phi[i] * f, (x, 0, 1))
rhs
```

```python
a = sym.zeros(3, 1)
for i in range(3):
    a[i] = rhs[i] / G[i,i]
a
```

```python
F = 0
for i in range(3):
    F = F + a[i] * phi[i]
F
```

```python
p1 = plot(f, (x, 0, 1), line_color='blue', show=False)
p2 = plot(F, (x, 0, 1), line_color='red', show=False)
p1.append(p2[0])
p1.show()
```
