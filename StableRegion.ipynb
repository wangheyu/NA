```python
import numpy as np
import matplotlib.pyplot as plt

# 定义参数θ从0到2π
theta = np.linspace(0, 2*np.pi, 400)
# 边界曲线参数方程
re_h = (1 - np.cos(theta))**2
im_h = np.sin(theta) * (2 - np.cos(theta))

# 生成复平面网格
x = np.linspace(-4, 5, 800)
y = np.linspace(-5, 5, 800)
X, Y = np.meshgrid(x, y)
H = X + 1j*Y

# 判断每个网格点是否稳定
stable = np.zeros_like(H, dtype=bool)
for i in range(len(x)):
    for j in range(len(y)):
        h = H[j, i]
        # 特征方程: (3 - 2h) z^2 - 4z + 1 = 0
        coeff = [3 - 2*h, -4, 1]  # 注意系数顺序: a*z^2 + b*z + c = 0
        roots = np.roots(coeff)
        if all(abs(root) < 1 for root in roots):
            stable[j, i] = True

# 绘图
plt.figure(figsize=(10, 6))
# 绘制稳定区域（用颜色填充）
plt.contourf(X, Y, stable, levels=[0.5, 1.5], colors=['lightblue'], alpha=0.6)
# 绘制边界曲线
plt.plot(re_h, im_h, 'r-', linewidth=2, label='Stability boundary')
plt.plot(re_h, -im_h, 'r-', linewidth=2)  # 对称部分
# 标注坐标轴和重要点
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.plot(0, 0, 'ko', label='Origin (0,0)')
plt.plot(4, 0, 'ro', label='Point (4,0)')
# 设置坐标轴范围
plt.xlim([-4, 5])
plt.ylim([-5, 5])
plt.xlabel(r'Re($\bar{h}$)')
plt.ylabel(r'Im($\bar{h}$)')
plt.title('Absolute Stability Region of BD2 Method')
plt.legend()
plt.grid(True, alpha=0.3)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
