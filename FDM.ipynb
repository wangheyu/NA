```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def source(x):
    return (np.pi)**2 * np.sin(np.pi * x)
N = 6
a = 0
b = 1
alpha = 0
beta = 0

h = (b - a) / N
x = np.linspace(a, b, N + 1)

A = np.zeros([N - 1, N - 1])
A[np.arange(N - 1), np.arange(N - 1)] = 2 * np.ones(N - 1) 
A[np.arange(N - 2), np.arange(N - 2) + 1] = -1 * np.ones(N - 2) 
A[np.arange(N - 2) + 1, np.arange(N - 2)] = -1 * np.ones(N - 2) 

b = h**2 * source(x[1:-1])
b[0] = b[0] + alpha
b[-1] = b[-1] + beta
u = np.zeros(N + 1)
u[1:-1] = np.linalg.solve(A, b)

plt.figure()
plt.plot(x, u, label='numerical')
plt.plot(x, np.sin(np.pi*x), 'r--', label='real')
plt.legend()
```
